# 0. Подключаем необходимые библиотеки и загружаем датасет

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_diabetes()
data, labels = dataset['data'], dataset['target']
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, random_state=0)
```

# 1. Построить график зависимости метрики качества классификации от количества деревьев

```python
scores = []
n_estimators = np.arange(1,120)

for i in n_estimators:
    clf = RandomForestClassifier(max_depth=10, random_state=0, n_estimators=i)
    clf.fit(X_train, Y_train)
    scores.append(clf.score(X_test,Y_test))

plt.plot(n_estimators, scores, color='purple')
plt.xlabel('Number of trees')
plt.ylabel('Accuracy')
plt.title('metrics: quality from the number of trees')
plt.show()
```

# 2. Построить график зависимости качества обучения от глубины деревьев

```python
scores = []
n_depths = np.arange(2, 40)

for i in n_depths:
    clf = RandomForestClassifier(max_depth = i, random_state=0)
    clf.fit(X_train, Y_train)
    scores.append(clf.score(X_test, Y_test))

plt.plot(n_depths, scores, color='purple')
plt.xlabel('Maximum depth')
plt.ylabel('Accuracy')
plt.title('The quality of learning on the depth of the trees')
plt.show()
```
